```
cities['center']=cities.geometry.centroid
```

```
cities.head(10)
```

```
part=['CITY','center']
city_centers=cities[part]
city_centers.head(5)
```

```
type(city_centers)
```

```
tampa_ctr=city_centers.center[2]
type(tampa_ctr)
```

```
tampa_city_center = [tampa_ctr.y, tampa_ctr.x]
print(tampa_city_center)
```

```
tampa_map=folium.Map(location=tampa_city_center, zoom_start=10, tiles='https://tile.thunderforest.com/spinal-map/{z}/{x}/{y}.png?apikey=c2d093a0cb4743f3869a1d2e5a814b02', attr="<a href=https://www.thunderforest.com/maps/>Spinal Map</a>")
```
